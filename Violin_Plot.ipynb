```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/JulioCastillo1/TP1-Datos/master/tweets_length.csv"

tweetsLengths = pd.read_csv(url)
tweetsLengths
```

```
data=tweetsLengths[['target','longitud']]
data
```

```
data.target.replace([0,1],['Falso','Verdadero'],inplace=True)
data
```

```

#plt.rc('xtick', labelsize=10)
#plt.rc('ytick', labelsize=10)
#plt.xlabel("Target", fontsize=20)
#plt.ylabel("Longitud de los tweets", fontsize=20)
#plt.figure(figsize=(30,20))
catplot=sns.catplot(x="target", y="longitud", kind="violin", inner=None, heigth=10, data=data)
plt.title("RelaciÃ³n entre la longitud de los tweets y la veracidad", fontsize=10)
#plt.figure(figsize=(30,20))
sns.swarmplot(x="target", y="longitud", size=2, data=data, ax=catplot.ax, hue="target")
plt.title("RelaciÃ³n entre la longitud de los tweets y la veracidad", fontsize=10)
#plt.figure(figsize=(30,20))
```
